# Penggabungan Dataset

### Menggabungkan history + achievement terlebih dahulu
Dilakukan penggabungan ini karena dataset achievement tidak memiliki variabel playerid. Digabungkan dengan dataset history karena kedua dataset ini berhubungan satu sama lain.

```python
hist_merge = hist.merge(achiev, on='achievementid', how='left')
hist_merge
```

### Agregasi ke level player, karena fokus analisis adalah analisis player. Dan itulah alasan variabel title dan description tidak digunakan karena kedua variabel tersebut tidak berkaitan dengan analisis players tetapi berkaitan dengan analisis games.

```python
df_hist = hist_merge.groupby('playerid').agg({
    'achievementid': 'count',
    'gameid': 'nunique',
    'rarity': 'nunique'
}).reset_index()

df_hist.rename(columns={
    'achievementid': 'total_achievement',
    'gameid': 'total_game_from_achievement',
    'rarity': 'variasi_rarity'
}, inplace=True)

df_hist
```

```python
df_hist.info()
```

# Penggabungan Dataset

```python
df1 = pd.merge(players, pg, on='playerid', how='left')
df1
```

```python
df1.info()
```

```python
df_final = pd.merge(df1, df_hist, on='playerid', how='left')
df_final
```

```python
df_final.info()
```

# Penanganan Missing Values

```python
df_final['total_achievement'] = df_final['total_achievement'].fillna(0)
```

```python
df_final['library'] = df_final['library'].fillna('No Games')
df_final['total_game_from_achievement'] = df_final['total_game_from_achievement'].fillna('No Achievement')
df_final['variasi_rarity'] = df_final['variasi_rarity'].fillna('No Achievement')
```
