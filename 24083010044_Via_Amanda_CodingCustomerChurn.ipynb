# Mengubah Tipe Data yang Tidak Sesuai

```python
data['SeniorCitizen'] = data['SeniorCitizen'].astype('category')
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

Stelah mengetahui, variabel mana yang memiliki tipe data yang tidak sesuai, maka perlu mengganti tipe data tersebut dengan syntax seperti di atas.

```python
data.info()
```

Setelah mengubah tipe data yang kurang sesuai dapat dilihat bahwa tipe data pada seniorcitezen dan totalcharges sudah berubah. Pada totalcharges memiliki nilai non-null 7032 karena setelah dirubah ke numerik, terdapat nilai yang tidak bisa dikonversi maka akan diubah menjadi null.

```python
summary = []

for col in data.columns:
    jumlah_unique = data[col].nunique()
    dtype = data[col].dtype

    if jumlah_unique == 2:
        jenis = "Biner"
    elif dtype in ['int64', 'float64']:
        jenis = "Numerik"
    else:
        jenis = "Kategorikal"

    summary.append([col, dtype, jumlah_unique, jenis])

summary_df = pd.DataFrame(summary, columns=["Variabel", "Dtype", "Jumlah_Unique", "Jenis_Variabel"])

summary_df
```

Kode diatas digunakan untuk mengklasifikasikan variabel. Jika memiliki jumlah nilai unik = 2 maka akan diklasifikasikan sebagai biner, dan jika tipe data numerik maka akan diklasifikasikan sebagai numerik. Kemudian hasil klasifikasi tersebut akan dibentuk dalam summary_df.

# Mengetahui Karakteristik Variabel Target

```python
freq = data['Churn'].value_counts()

prop = data['Churn'].value_counts(normalize=True) * 100

print("Frekuensi:\n", freq)
print("\nPersentase (%):\n", prop)
```

Kode di atas digunakan untuk menghitung jumlah pelanggan yang berlangganan dan tidak berlangganan, serta menghitung persentasenya. Dapat dilihat pada output, pelanggan banyak memilih berhenti berlangganan dibandingkan dengan pelangganan yang berlangganan, dengan persentase sebesar 73% untuk yang tidak berlangganan dan 26% untuk yang berlangganan.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Churn', data=data)

plt.title('Distribusi Kelas Churn')
plt.xlabel('Churn')
plt.ylabel('Jumlah Pelanggan')
plt.show()
```

Kode di atas digunakan untuk memvisualisasikan seberapa banyak pelanggan yang berlangganan dan tidak berlangganan.

# Mengetahui Hubungan Variabel Target denan Variabel Input

```python
sns.countplot(x='Contract', hue='Churn', data=data)

plt.title('Distribusi Churn berdasarkan Contract')
plt.xticks(rotation=0)
plt.show()
```

Kode di atas digunakan untuk visualisasi variabel target dengan variabel input kategorik.

```python
categorical_cols = ['Contract', 'InternetService', 'PaymentMethod',
                    'OnlineSecurity', 'TechSupport']

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue='Churn', data=data)
    plt.title(f'Distribusi Churn berdasarkan {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=data)

plt.title('MonthlyCharges berdasarkan Churn')
plt.show()
```

Kode di atas digunakan untuk visualisasi variabel target dengan variabel input yang numerik, seperti tenure, monthlycharges, dan totalcharges.

```python
numerical_cols = data.select_dtypes(include=['int64','float64']).columns.tolist()

for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Churn', y=col, data=data)
    plt.title(f'{col} berdasarkan Churn')
    plt.tight_layout()
    plt.show()
```
